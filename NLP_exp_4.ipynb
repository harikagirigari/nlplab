```python
import pandas as pd
import numpy as np
import re
```

```python
data_out = pd.read_csv('/content/twitter4000.csv')
```

```python
#For removing emails
pattern = r'\w+@\w+.\w+'
```

```python
pattern = r'\b[A-za-z._+-]+@[A-za-z.].[A-z|a-z]{2,}\b'
```

```python
data_out['twitts'].apply(lambda x: re.findall(pattern=pattern, string=x))
```

```python
#For removing URLs
pattern = r'http\S+|www\S+'
data_out['twitts'].apply(lambda x: re.findall(pattern=pattern, string=x))
```

```python
#For removing Numbers
pattern = r'\w+[0-9]\w+'
data_out['twitts'].apply(lambda x: re.findall(pattern=pattern, string=x))
```
