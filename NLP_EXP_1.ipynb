```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
input ="Hi I am Harika. How are you"
output1 = word_tokenize(input)
print(output1)
```

```python
output2 = sent_tokenize(input)
print(output2)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
input = "This is second session"
doc = nlp(input)
```

```python
for token in doc:
   print(token)
```

```python
for sent in doc.sents:
  print(sent)
```
